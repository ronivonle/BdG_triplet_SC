# LDOS

```julia
figure(figsize=(10,8))
f = open("corner.dat")
s = readdlm(f)
s = reshape(s,(convert(Int32,length(s)/3),3))
x = s[:,1]
y = s[:,2]
z = s[:,3]
a1 = scatter(x,y,z*20,c=z,edgecolors="b",cmap="Reds")
colorbar(a1)
xticks(0:5:maximum(x))
yticks(0:5:maximum(y))
title("CornerState")
savefig("cor.eps",bbox_inches="tight",dpi=300)
```

# Bilayer system

## LDOS for system

```julia
# SC LDOS
figure(figsize=(10,8))
f = open("scldos.dat")
s = readdlm(f)
s = reshape(s,(convert(Int32,length(s)/3),3))
x = s[:,1]
y = s[:,2]
z = s[:,3]
a1 = scatter(x,y,z*20,c=z*0.1,edgecolors="b",cmap="Reds")
colorbar(a1)
xticks(0:5:maximum(x))
yticks(0:5:maximum(y))
xlabel("X")
ylabel("Y")
title("local density of state in sc layer")
savefig("scrho.eps",bbox_inches="tight",dpi=300)
#  TI LDOS
figure(figsize=(10,8))
f = open("tildos.dat")
s = readdlm(f)
s = reshape(s,(convert(Int32,length(s)/3),3))
x = s[:,1]
y = s[:,2]
z = s[:,3]
a1 = scatter(x,y,z*20,c=z*0.1,edgecolors="b",cmap="Reds")
colorbar(a1)
xticks(0:5:maximum(x))
yticks(0:5:maximum(y))
xlabel("X")
ylabel("Y")
title("local density of state in tilayer")
savefig("tirho.eps",bbox_inches="tight",dpi=300)
```

## Energy band

```julia
figure(figsize=(10,8))
f = open("engband.dat")
s = readdlm(f)
#s = reshape(s,(convert(Int32,length(s)/3),3))
p=plot(s[:,1],s[:,2:end],"black")
title("d-wave")
xlabel("kx")
ylabel("ky")
annotate("(a)", xy=(0,0), xytext=(-0.3, 3.5))
savefig("eband.eps",bbox_inches="tight",dpi=300)
close(f)
```

```julia
figure(figsize=(10,8))
f = open("bilayerband.dat")
s = readdlm(f)
#s = reshape(s,(convert(Int32,length(s)/3),3))
p=plot(s[:,1],s[:,2:end],"black",linewidth=0.5,s[:,1],s[:,3000],"r",s[:,1],s[:,3001],"b")
title("bilayer")
xlabel("kx")
ylabel("ky")
xlim(-1,1)
ylim(-0.2,0.2)
#annotate("(a)", xy=(0,0), xytext=(-0.3, 3.5))
savefig("biband.eps",bbox_inches="tight",dpi=300)
close(f)
```
